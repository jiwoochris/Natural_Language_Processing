# NUMPY 실습

```python
import numpy as np
import pandas as pd
import torch
```

Ndarray  
numpy.array()함수에 list like형태의 데이터를 입력하여 생성한다.  
doc: https://compmath.korea.ac.kr/appmath/NumpyBasics.html  

```python
data = [1,2,3]
print(type(data))

numpy_data = np.array(data)
print(numpy_data)
print(type(numpy_data))
```

Ndarray는 행렬 형태의 데이터   
행렬은 각 행과 열에 데이터의 크기가 같아야한다.

python의 list는 단순 데이터의 묶음 형태이기 때문에 [[1,2],1,(1,2,3)] 같이 각 element의 형태, 크기, 타입이 자유로움   
numpy의 ndarray는 행렬, 텐서와 같은형태를 가지기 때문에 각 elements의 타입이 동일해야하며 모든 열과 행의 크기가 같아야 함

```python
python_list = [[1,2],1,(1.0,2,3),"string"]
print(python_list)

#ndarray는 모든 행과 열의 크기가 같아야하기에 정상적인 데이터가 생성되지 않음
a = [[1,2],[1,2,3]]
np_a = np.array(a)
print("\n")
print(np_a)
print(np_a.shape)

#2차원 list를 이용한 ndarray 생성
b = [[1,2,3],[4,5,6]]
np_b = np.array(b)
print("\n")
print(np_b)
print(np_b.shape)
```

ndarray는 상수와의 연산, ndarray와의 연산, 행렬연산 등 다양한 연산을 지원한다.

```python
b = [[1,2,3],[4,5,6]]
np_b = np.array(b)

c = [[6,5,4],[3,2,1]]
np_c = np.array(c)

print("tensor b")
print(np_b)
print(np_b.shape)

print("\ntensor c")
print(np_c)
print(np_c.shape)

print("\ntranspose c")
print(np.transpose(np_c))
print(np_c.T)

print("\n flatten c")
print(np_c.flatten())

print("\nreshape c")
print(np_c.reshape((3,2)))
print(np_c.shape)

print("\nb*2:")
print(np_b*2)

#element wise add element단위 덧셈
print("\nb+c:")
print(np_b+np_c)

#element wise product element단위 곱셈
print("\nb*c:")
print(np_b*np_c)

#dot 행렬곱
print("\nb dot transpose c")
print(np.dot(np_b,np_c.T))
```
