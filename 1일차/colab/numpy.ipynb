Ndarray  
numpy.array()함수에 list like형태의 데이터를 입력하여 생성한다.  
doc: https://compmath.korea.ac.kr/appmath/NumpyBasics.html  

```python
data = [1,2,3]
print(type(data))

np_array = np.array(data)
print(type(np_array))
```

Ndarray는 행렬 형태의 데이터   
행렬은 각 행과 열에 데이터의 크기가 같아야한다.

python의 list는 단순 데이터의 묶음 형태이기 때문에 [[1,2],1,(1,2,3)] 같이 각 element의 형태, 크기, 타입이 자유로움   
numpy의 ndarray는 행렬, 텐서와 같은형태를 가지기 때문에 각 elements의 타입이 동일해야하며 모든 열과 행의 크기가 같아야 함

```python
l = [[1,2],1,(1,2,3), 'str']
np_array = np.array(l) # error

l = [[1,2,3], [1,2,3]]
a = np.array(l)
print(a.shape)
```
