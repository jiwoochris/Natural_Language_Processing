#Linear Regresion

여러 방법론을 이용한 Linaer Regression Optimization

Dataset load  
data link: https://www.kaggle.com/datasets/andonians/random-linear-regression?resource=download  
csv(comma-separated values) 콤마로 각 데이터가 구분하도록 데이터를 저장한 txt파일

```python
import pandas as pd
import os

data_path = "/content/drive/MyDrive/랩세미나(new)/tutorial/open_tutorial_DL_NLP/1일차/colab/LR_dataset"
train_file = "train.csv"
test_file = "test.csv"

df_data = pd.read_csv(os.path.join(data_path, train_file)).dropna()
print(df_data)

data_list_x = list(df_data.loc[:,"x"])
data_list_y = list(df_data.loc[:,"y"])
data_num = len(data_list_x)
print(len(data_list_x))
print(len(data_list_y))
```

ML hypothesis와 loss function설정  
hypo: y = a * x + b  
loss: MSE loss (y-y_pred) ^ 2

```python
def lr_hypo(a,b,x):
  return a * x + b

def lr_loss(y, y_pred):
  return (y-y_pred) * (y-y_pred)
```

Least Square Method

```python
import numpy as np

print(data_list_x)
mean_x = np.mean(data_list_x)
print(data_list_y)
mean_y = np.mean(data_list_y)
print(mean_x)
print(mean_y)
tmp_top = 0
tmp_bottom = 0
for i in range(data_num):
  tmp_top += (data_list_y[i]-mean_y)*(data_list_x[i]-mean_x)
  tmp_bottom += (data_list_x[i]-mean_x)*(data_list_x[i]-mean_x)

a = tmp_top/tmp_bottom
b = mean_y - a * mean_x

print("a = ",a)
print("b = ",b)

loss_sum = 0
for i in range(data_num):
  pred_y = lr_hypo(a,b,data_list_x[i])
  loss_sum += lr_loss(data_list_y[i], pred_y)

loss_mean = loss_sum/data_num
print(loss_mean)
```

최대한 많은 case에서 loss 구해보기

```python
import json

a_list = np.arange(0,2,0.1)
print(a_list)
b_list = np.arange(-1,1,0.1)
print(b_list)

result_dict = {}



for a_i in range(len(a_list)):
  a = a_list[a_i]
  result_dict[a] = {}
  for b_j in range(len(b_list)):
    b = b_list[b_j]
    loss_sum=0
    for i in range(data_num):
      pred_y = lr_hypo(a,b,data_list_x[i])
      loss_sum += lr_loss(data_list_y[i], pred_y)
    loss_mean = loss_sum/data_num
    result_dict[a][b] = loss_mean

print(json.dumps(result_dict,indent=2))
```

gradient descent

MSE의 미분 if f(x)가 hypo일때  
MSE: (f(x) - y) ^ 2 / n  
MSE의 편미분: 2 * (f(x) - y)f'(x) / n

```python
def ff_a(a,b,x,y):
  return 2*(a*x+b-y)*x

def ff_b(a,b,x,y):
  return 2*(a*x+b-y)

# from sklearn.preprocessing import RobustScaler
# data_list_x = RobustScaler().fit_transform(np.array(data_list_x).reshape(-1,1)).reshape(-1)
# print(data_list_x)
print(data_list_x)


a = 10
b = -1
lr = 0.0001
for e in range(100):
  print("epoch ",e)
  loss_sum = 0
  grad_a_sum = 0
  grad_b_sum = 0
  
  print(a,b)
  for i in range(data_num):
    pred_y = lr_hypo(a,b,data_list_x[i])
    loss_sum += lr_loss(data_list_y[i], pred_y)
    grad_a_sum += ff_a(a,b,data_list_x[i],data_list_y[i])
    grad_b_sum += ff_b(a,b,data_list_x[i],data_list_y[i])
  grad_a_mean = grad_a_sum/data_num
  grad_b_mean = grad_b_sum/data_num
  loss_mean = loss_sum/data_num
  a = a - lr * grad_a_mean
  b = b - lr * grad_b_mean

  print(loss_mean)
  print("\n\n\n")
```

#Multiple Linear Regression


input으로 여러개의 입력이 들어가는 LR  
dataset: https://www.kaggle.com/datasets/fayejavad/marketing-linear-multiple-regression  
x = ["youtube","facebook","newspaper"]  
y = ["sales"]

```python
data_path = "/content/drive/MyDrive/랩세미나(new)/tutorial/open_tutorial_DL_NLP/1일차/colab/LR_dataset"
train_file = "Marketing_Data.csv"

multi_df_data = pd.read_csv(os.path.join(data_path,train_file))
print(multi_df_data)

x_multi_df = multi_df_data.loc[:, ["youtube","facebook","newspaper"]]
print(x_multi_df)
y_sales = list(multi_df_data.loc[:, "sales"])
print(y_sales)

x_youtube = list(x_multi_df.loc[:, "youtube"])
print(x_youtube)
x_facebook = list(x_multi_df.loc[:, "facebook"])
print(x_facebook)
x_newspaper = list(x_multi_df.loc[:, "newspaper"])
print(x_newspaper)

# from sklearn.preprocessing import RobustScaler
# x_multi_df = RobustScaler().fit_transform(x_multi_df)

# print(x_multi_df)
```

x가 두개 들어가므로  
hypo: ax1 + bx2 + cx3+ d = y  
loss function을 a,b,c에 대해 서 편미분  
MSE: (f(x) - y) ^ 2 / n  
MSE의 편미분: 2 * (f(x) - y)f'(x) / n

```python
def f(a,b,c,d,x1,x2,x3):
  return a * x1 + b * x2 + c * x3 + d

def loss(y,y_pred):
  return (y_pred - y) * (y_pred - y)

def ffa(a,b,c,d,x1,x2,x3,y):
  return 2 * (a * x1 + b * x2 + c * x3 + d - y)*x1

def ffb(a,b,c,d,x1,x2,x3,y):
  return 2 * (a * x1 + b * x2 + c * x3 + d - y)*x2

def ffc(a,b,c,d,x1,x2,x3,y):
  return 2 * (a * x1 + b * x2 + c * x3 + d - y)*x3

def ffd(a,b,c,d,x1,x2,x3,y):
  return 2 * (a * x1 + b * x2 + c * x3 + d - y)

data_num = len(multi_df_data)
a = 10
b = 3
c = 10
d = 5
lr = 0.00002
for e in range(100):
  print("epoch ",e)
  loss_sum = 0
  grad_a_sum = 0
  grad_b_sum = 0
  grad_c_sum = 0
  grad_d_sum = 0
  
  print(a,b,c,d)
  for i in range(len(multi_df_data)):
    pred_y = f(a,b,c,d,x_youtube[i], x_facebook[i], x_newspaper[i])
    loss_sum += loss(y_sales[i], pred_y)
    grad_a_sum += ffa(a,b,c,d,x_youtube[i], x_facebook[i], x_newspaper[i],y_sales[i])
    grad_b_sum += ffb(a,b,c,d,x_youtube[i], x_facebook[i], x_newspaper[i],y_sales[i])
    grad_c_sum += ffc(a,b,c,d,x_youtube[i], x_facebook[i], x_newspaper[i],y_sales[i])
    grad_d_sum += ffd(a,b,c,d,x_youtube[i], x_facebook[i], x_newspaper[i],y_sales[i])
  grad_a_mean = grad_a_sum/data_num
  grad_b_mean = grad_b_sum/data_num
  grad_c_mean = grad_c_sum/data_num
  grad_d_mean = grad_d_sum/data_num
  loss_mean = loss_sum/data_num
  a = a - lr * grad_a_mean
  b = b - lr * grad_b_mean
  c = c - lr * grad_c_mean
  d = d - lr * grad_d_mean

  print(loss_mean)
  print("\n\n\n")
```
